Columns to clean:
1. Convert values in Height column to 1 scale(cm)
2. Convert values in Weight column to 1 scale(kgs)
3. Remove new line character in Club column
4. Create Day, Month Year columns based on Joined column
5. Clean Value, Wage and Release_Clause columns
6. Some columns have 'star' characters. Strip those columns of these stars and make the columns numerical

```python
# df['Club']
```

```python
converted_club=[]

for clubs in df['Club']:
    clubs=clubs.replace("\n",'')
    converted_club.append(clubs)
df['Club']=converted_club
# df['Club']
```

```python
converted_heights=[]

for height in df['Height']:
    if "'" in height or '"' in height:
        feet, inches=height.split("'")
        inches=inches.strip('"')
        total_inches=int(feet)*12+int(inches)
        cm=total_inches*2.54
        converted_heights.append(cm)
    else:
        cm = float(height[:-2])  # Remove the last two characters ("cm")
        converted_heights.append(cm)
```

```python
df['Height']=converted_heights
# df['Height']
```

```python
# for weight in df['Weight']:
#     print(weight)
```

```python
weight_to_kgs=[]

for weight in df['Weight']:
    if 'lbs' in weight:
        lbs = float(weight[:-3])
        weight_kg=lbs*0.4535
        weight_to_kgs.append(weight_kg)
    else:
        kg=float(weight[:-2])
        weight_to_kgs.append(kg)
        
# print(weight_to_kgs)

df['Weight']=weight_to_kgs
# df['Weight']    
```

```python
# df['Joined']
```

```python
from datetime import datetime

df['Joined']=pd.to_datetime(df['Joined'], format="%b %d, %Y")
# df['Joined']
```

```python
df['Year']=df['Joined'].dt.year
df['Month']=df['Joined'].dt.month
df['Day']=df['Joined'].dt.day

df.head()

```

```python
# for i in df['Value']:
#     print(i)
```

```python
converted_value=[]

for value in df['Value']:
    value=value.replace('M','000000')
    value=value.replace('K','000')
    value = value.replace('.', '')  # Remove .
    value=value.replace("€",'')
    converted_value.append(int(value))
    
df['Value']=converted_value
# df['Value']
```

```python
# for i in df['Wage']:
#     print(i)
```

```python
converted_wage=[]

for wage in df['Wage']:
    wage=wage.replace('K','000')
    wage=wage.replace('€','')
    converted_wage.append(int(wage))
    
df['Wage']=converted_wage
# df['Wage']

```

```python
# for i in df['Release Clause']:
#     print(i)
```

```python
converted_clause=[]

for clause in df['Release Clause']:
    clause=clause.replace('€','')
    clause=clause.replace('.','')
    clause=clause.replace('M','000000')
    clause=clause.replace('K','000')
    converted_clause.append(int(clause))

df['Release Clause']=converted_clause
# df['Release Clause']
```

```python
converted_hits=[]

for hit in df['Hits']:
    if pd.notnull(hit):  # Check if the value is not NaN
        if isinstance(hit, float):
            hit = str(hit).replace('K', '000').replace('.', '')
        else:
            hit = hit.replace('K', '000').replace('.', '')
        converted_hits.append(int(hit))
    
# Create a new Series object with converted_hits
hits_series = pd.Series(converted_hits)

# Assign the Series to the 'Hits' column of the DataFrame
df['Hits'] = hits_series

# print(df['Hits'])
```

```python
df['Hits'].describe()
```

```python
import numpy as np

median_hits=np.nanmedian(df['Hits']) # Calculate the median ignoring NaN values

df['Hits'] = df['Hits'].fillna(median_hits) # Impute missing values with the median
# df['Hits'].describe()
```

```python
#As Hits column has outliers and mean is effected by outliers, imputed null values with median
```
